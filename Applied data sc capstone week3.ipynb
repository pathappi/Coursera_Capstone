```python
import pandas as pd
from urllib.request import urlopen
# import beautifulsoup for data scrapping
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
type(soup)
# extract cell data
cells = soup.find('table',{'rules':'all'}).find_all('p')
```

```python
postal_codes = []
boroughs = []
neighborhoods = []
for cell in cells:
    if not cell.find('i'):
        content = cell.get_text()
        postal_codes.append(content[0:3])
        content = content[3:] # extract first 3 char as postal code
        content = content.replace('/', ',') # replace / by , for neighborhoods
        if '(' in content: # if borough exists
            actual_borough = content.split('(', 1)[0]
            boroughs.append(actual_borough)
            char_to_remove = [actual_borough, '(', ')', '\n']
            new_char = content
            for char in char_to_remove:
                new_char = new_char.replace(char, '')
            neighborhoods.append(new_char)   
        else: # if borough doesnt exist
            boroughs.append('None')
            char_to_remove = ['(', ')', '\n']
            new_char = content
            for char in char_to_remove:
                new_char = new_char.replace(char, '')
            neighborhoods.append(new_char)
     
data = {'PostalCode':postal_codes,'Borough':boroughs,'Neighborhood':neighborhoods} # data format
df = pd.DataFrame(data) # convert to dataframe
df
```

```python
postal_codes_data = pd.read_csv("http://cocl.us/Geospatial_data") # extract postal codes coordinates csv

# pull longitude and latitude from csv and add to main dataframe
for index_pc, row_pc in df.iterrows():
    for index, row in postal_codes_data.iterrows(): # loop in csv to search coordinates
        if row_pc['PostalCode'] == row['Postal Code']:
            df.loc[index_pc, 'Latitude'] = row['Latitude']
            df.loc[index_pc, 'Longitude'] = row['Longitude']
df
```
